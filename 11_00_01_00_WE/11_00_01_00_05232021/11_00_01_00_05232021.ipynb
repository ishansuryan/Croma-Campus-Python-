```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib_inline
```

```python
iris = load_iris()
x = iris['data'][:,2:]
y = iris['target']
```

```python
iris.keys()
```

```python
dt_clf = DecisionTreeClassifier(max_depth = 2,  criterion = 'entropy')
```

```python
dt_clf.fit(x,y)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(7,7))
plot_tree(dt_clf,feature_names= iris['feature_names'][2:],class_names= iris['target_names'],filled = True)
plt.show()
```

```python
# ins = [[2,6]]
pred_prob = dt_clf.predict_proba([[2,6]])
pred_prob
```
