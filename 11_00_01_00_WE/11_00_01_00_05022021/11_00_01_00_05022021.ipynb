```python
from sklearn import datasets
import numpy as np
np.random.seed(42)
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
iris['target_names']
```

```python
iris['feature_names']
```

```python
x = iris['data'][:,3:]
y = (iris['target']== 2) # for iris verginica will return True
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_clf = LogisticRegression()
```

```python
log_clf.fit(x,y)
```

```python
x_new = np.linspace(0,3,50).reshape(-1,1)
y_pred = log_clf.predict_proba(x_new)
```

```python
not_verginica = (iris['target'] != 2) 
```

```python
not_verginica = iris['data'][not_verginica]
```

```python
verginica = iris['data'][y]
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_new,y_pred[:,0],'k--', label = 'Not iris Verginica')
plt.plot(x_new,y_pred[:,1],'g--', label = 'Iris-Verginica')
plt.plot(not_verginica[:,3:],np.zeros((len(not_verginica),1)), 'bs')
plt.plot(verginica[:,3:], np.ones((len(verginica),1)), 'g^')
plt.xlabel('petal width(cm)')
plt.ylabel('Probalility')
plt.legend()
plt.grid()
plt.show()
```
