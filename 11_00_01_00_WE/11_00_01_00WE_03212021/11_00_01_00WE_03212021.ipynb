```python
import numpy as np
```

```python
x = 6 * np.random.rand(100,1) - 3
```

```python
y = 2*x**2 + 2*x -3 + 3*np.random.rand(100,1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'b.')
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', rotation = 0, fontsize = 15)
plt.show()
```

```python
x[0]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_featires_2 = PolynomialFeatures(degree=2, include_bias=False)
```

```python
x_poly_2 = poly_featires_2.fit_transform(x)
```

```python
x_poly_2[0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_poly_2, y)
```

```python
lin_reg.intercept_, lin_reg.coef_
```

```python
x_test = np.linspace(-3,3,100).reshape(100,1)
x_test_poly = poly_featires_2.fit_transform(x_test)
```

```python
y_pred = lin_reg.predict(x_test_poly)
```

```python
plt.plot(x_test,y_pred,'r-')
plt.plot(x,y,'b.')
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', rotation = 0, fontsize = 15)
plt.show()
```
