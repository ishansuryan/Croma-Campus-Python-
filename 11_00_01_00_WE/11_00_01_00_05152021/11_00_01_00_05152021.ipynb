```python
import numpy as np
import matplotlib.pyplot as plt
# % matplotlib_inline
```

```python
x = np.array([[1, 50], [5, 20], [3, 80], [5,60]])
y = np.array([0,0,1,1])
```

```python
from sklearn.svm import SVC
svm_clf = SVC(C = 100 ,kernel = 'linear')
svm_clf_scaled = SVC(C = 100 ,kernel = 'linear')
svm_clf.fit(x,y)
```

```python
svm_clf.coef_,svm_clf.intercept_
```

```python
svm_clf.coef_[0]
```

```python
def plot_svc_db(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]

    # At decision boundry w0x0 + w1x1 + b = 0
    # x1 = -w0x0/w1 -b/w1

    x0 = np.linspace(xmin,xmax,200)
    decision_boundry = -w[0]/w[1] * x0 -  b/w[1]

    margin = 1/w[1]
    line_up = decision_boundry + margin
    line_down = decision_boundry - margin

    svs = svm_clf.support_vectors_
    plt.scatter(svs[:,0],svs[:,1], s = 180)
    plt.plot(x0, decision_boundry, 'k-', lw = 3)
    plt.plot(x0, line_up, 'k--', lw = 2)
    plt.plot(x0, line_down, 'k-', lw = 2)
```

```python
plt.plot(x[:,0][y==1], x[:,1][y==1],'bo')
plt.plot(x[:,0][y==0], x[:,1][y==0],'ms')
plot_svc_db(svm_clf,0,5.5)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
svm_clf_scaled.fit(x_scaled,y)
plt.plot(x_scaled[:,0][y==1], x_scaled[:,1][y==1],'bo')
plt.plot(x_scaled[:,0][y==0], x_scaled[:,1][y==0],'ms')
plot_svc_db(svm_clf_scaled,-1.5,1.5)
plt.show()
```
