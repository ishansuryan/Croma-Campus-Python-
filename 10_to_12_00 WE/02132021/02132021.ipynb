```python
def line(m,x,c=0):
    y = m * x + c
    return y
```

```python
import matplotlib.pyplot as plt
```

```python
m = 1
x1,x2  = 2, 6
y1, y2 = line(m,x1), line(m,x2)
x_num = [0,x1,x2]
y_num = [0,y1,y2]
```

```python
plt.plot(x_num, y_num, marker = '*')
plt.xlim((-10,10))
plt.ylim((-10,10))
plt.show()
```

```python
m = 2
c = 1
x1,x2  = 2, 6
y1, y2 = line(m,x1,c), line(m,x2,c)
x_num = [x1,x2]
y_num = [y1,y2]
```

```python
plt.plot(x_num, y_num, marker = '*')
plt.xlim((0,30))
plt.ylim((0,30))
plt.show()
```
