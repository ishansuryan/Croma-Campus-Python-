```python
np.random.seed(42)
X = 2 * np.random.rand(100,1)
```

```python
y = 4 + 3 * X + np.random.randn(100,1)
#               Gaussian Noise
```

```python
plt.plot(X,y,'b.')
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X]
```

```python
%timeit theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
```

```python
y_predict = X_new_b.dot(theta)
y_predict
```

```python
plt.plot(X_new,y_predict,'r-')
plt.plot(X,y,'b.')
plt.xlabel('$X_1$', fontsize = 20)
plt.ylabel('y', fontsize = 20)
plt.axis([0,2,0,15])
plt.show()
```
