```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
x = pd.Series(np.random.randint(10, 25, 30)) 
```

```python
l = list(x.mode())
```

```python
sns.kdeplot(x, shade=True, color='green') 
plt.plot([x.mean(),x.mean()], [0,0.13], 'k-',label='Mean') 
plt.plot([x.median(),x.median()], [0,0.13], 'g--',label='Median') 
for i in range(len(x.mode())): 
    plt.plot([x.mode()[i],x.mode()[i]], [0,0.13], 'r-.', label = f'mode{i+1}') 
plt.legend() 
# if x.mean()==x.median(): plt.title('Symmetrical') else: plt.title('Right Skewed' if x.mean()>x.median() else 'Left Skewed') 
plt.show()
```
