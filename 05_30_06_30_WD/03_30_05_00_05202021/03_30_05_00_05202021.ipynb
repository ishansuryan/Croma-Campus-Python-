```python
def invalid(l):
    r = []
    for i in list(l):
        try:
            i = float(i)
        except :
            i = np.nan
        r.append(i)
    return r
```

```python
lst = ['s',1,'t',3,5, 'p']
z = invalid(lst)
list(z)
```
