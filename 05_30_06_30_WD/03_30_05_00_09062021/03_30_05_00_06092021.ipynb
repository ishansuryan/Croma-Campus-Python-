```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
fitness = pd.read_csv('/mnt/New_Volume/Work_From_Home/Data/CardioGoodFitness.csv')
fitness.head()
```

```python
# plotting a histogram from pandas
fitness['Age'].hist(figsize = (9,6), bins = 33)
plt.xlabel('Age', size= 16)
plt.ylabel('Count', size = 16)
plt.title('Histogram of Age')
plt.xticks([x for x in range(18,51)])
plt.yticks(np.arange(26))
plt.axis([18,51,0,26])
plt.show()
```

```python
plt.figure(figsize=(20,12))
sns.histplot(data = fitness, x = 'Age')
plt.title('Histogram of Age')
plt.xticks([x for x in range(18,51)])
plt.yticks(np.arange(50))
plt.axis([18,51,0,50])
plt.show()
```

```python
sns.distplot(fitness['Age'])
plt.title('Histogram of Age')
# plt.xticks([x for x in range(18,51)])
# plt.yticks(np.arange(26))
# plt.axis([18,51,0,26])
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.hist(fitness['Age'], bins = 33)
plt.plot([26,26], [0,50], 'r--', label = 'Median',linewidth = 10)
plt.plot([28.788889,28.788889],[0,50],'k-',label = 'Mean',lw=6)
plt.xlabel('Age', size= 16)
plt.ylabel('Count', size = 16)
plt.title('Histogram of Age')
plt.xticks([x for x in range(18,51)])
plt.yticks(np.arange(26))
plt.axis([18,51,0,26])
plt.grid(True)
plt.legend()
plt.show()
```
