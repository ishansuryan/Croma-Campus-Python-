```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.arange(0,8)
k
```

```python
n = 7
p = 0.6
```

```python
prob_dist = stats.binom.pmf(k,n,p)
```

```python
prob_dist
```

```python
prob_dist.sum()
```

```python
plt.plot(prob_dist, marker = '*')
plt.grid()
plt.xlabel('Number of Customer pay on Time',size = 20 )
plt.ylabel('Probalility',size = 20)
plt.show()
```
