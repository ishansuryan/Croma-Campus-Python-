```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 4*np.random.rand(100).reshape(-1,1) - 2
```

```python
y = 2*x -3 + 5* np.random.rand(100).reshape(-1,1)
```

```python
plt.plot(x.reshape(100,-1),y.reshape(100,-1),'bo')
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x,y)
```

```python
lin_reg.coef_,lin_reg.intercept_
```

```python
yticks = [x for x in range(-6,6,2)]
yticks.extend([-0.4,0.4])
yticks
```

```python
x_new = np.linspace(-2,2,200).reshape(-1,1)
y_new = lin_reg.predict(x_new)
plt.figure(figsize = (10,10))
plt.plot(x.reshape(100,-1),y.reshape(100,-1),'bo')
plt.plot(x_new,y_new,'r-')
plt.yticks(yticks)
plt.grid()
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd_reg = SGDRegressor(random_state=42)
```

```python
sgd_reg.fit(x,y.ravel())
```

```python
sgd_reg.coef_, sgd_reg.intercept_
```
