```python
import pandas as pd
```

```python
data = pd.read_csv('/mnt/New_Volume/Work_From_Home/Data/SBIN.csv')
```

```python
data.columns = [x.strip() for x in data.columns]
```

```python
data.shape  
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['OPEN'],data['close'],'b*')
```

```python
x = data[['OPEN','VOLUME']]
y = data['close']
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x,y)
```

```python
x_new_open = np.linspace(x['OPEN'].min(),x['OPEN'].max(),100)
x_new_volume = np.linspace(x['VOLUME'].min(),x['VOLUME'].max(),100)
```

```python
x_new = pd.concat([pd.Series(x_new_open), pd.Series(x_new_volume)], axis = 1, ignore_index=True)
x_new.shape
```

```python
y_pred = lin_reg.predict(x_new)
```

```python
plt.plot(data['OPEN'],data['close'],'bo')
plt.plot(x_new_open,y_pred,'r-')
```

```python
plt.plot(data['VOLUME'],data['close'],'bo')
plt.plot(x_new_volume,y_pred,'r-')
```
