```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 4*np.random.rand(100,1) -2
y = 3*x**2 - 0.8*x + 3.2 + 5*np.random.rand(100,1)
```

```python
plt.plot(x,y,'bo')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_2 = PolynomialFeatures(degree=2, include_bias = False)
```

```python
x_tr = poly_2.fit_transform(x)
# x_tr
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_tr,y)
```

```python
x_new = np.linspace(-2,2,200).reshape(200,1)
x_new_poly = poly_2.fit_transform(x_new)
y_pred = lin_reg.predict(x_new_poly)
```

```python
plt.plot(x,y,'bo')
plt.plot(x_new,y_pred,'r-')
```

```python
poly_300 = PolynomialFeatures(degree= 30, include_bias=False)
```

```python
x_300 = poly_300.fit_transform(x)
# len(x_300[0])
```

```python
lin_reg_300 = LinearRegression()
lin_reg_300.fit(x_300,y)
x_new_300 = poly_300.fit_transform(x_new)
y_300_pred = lin_reg_300.predict(x_new_300)
plt.plot(x,y,'bo')
plt.plot(x_new,y_300_pred,'r-')
plt.ylim(-5,20)
```
