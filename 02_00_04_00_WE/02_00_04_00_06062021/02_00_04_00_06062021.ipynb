```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# import matplotlib
%matplotlib inline
# plt.rcParams["figure.figsize"] = (20,20)
# matplotlib.rcdefaults()
```

```python
fitness = pd.read_csv('/mnt/New_Volume/Work_From_Home/Data/CardioGoodFitness.csv')
fitness.head()
```

```python
# plt.figure(figsize=(10,10))
fitness.hist(figsize = (20,10))
plt.show()
```

```python
fitness['Age'].hist(figsize = (10,6), grid= False,color = 'brown',)
plt.xlabel ('Age', size = 20)
plt.ylabel('Frequency',size = 20)
plt.title('Age Histogram')
plt.show()
```

```python
plt.figure(figsize=(10,6))
# plt.grid()
sns.histplot(data = fitness, x = 'Age',bins= 32)
plt.show()
```

```python
fitness.describe()
```

```python
fitness.plot(kind= 'box',y = 'Age',figsize = (10,6), meanline = True, showmeans = True)
# plt.yticks([x for x in range(17,51,1)])
plt.grid()
plt.plot()
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = 'Gender', y = 'Age', data = fitness, showmeans = True, meanline = True)
plt.yticks([x for x in range(16,51,2)])
plt.grid()
plt.show()
```
