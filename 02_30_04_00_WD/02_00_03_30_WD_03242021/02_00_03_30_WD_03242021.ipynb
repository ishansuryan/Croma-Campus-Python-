```python
import numpy as np
```

```python
def leaky_relu(z,a = 0.2 ):
    return np.maximum(a*z,z)
```

```python
x = np.linspace(-10,10,200)
y = leaky_relu(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'r-')
plt.xticks([x for x in range(-10,11)])
plt.yticks([x for x in range(-5,11)])
plt.grid()
plt.show()
```
