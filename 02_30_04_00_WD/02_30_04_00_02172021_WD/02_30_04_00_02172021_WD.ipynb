```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
x = 2* np.random.rand(100,1)
```

```python
y = 4+3 *x + np.random.randn(100,1)
```

```python
plt.plot(x,y,'b.')
plt.xlabel("$X_1$", fontsize = 20)
plt.ylabel('y', rotation =0, fontsize = 20)
plt.axis([0,2.5,0,15])
plt.show()
```

```python
x_b = np.c_[np.ones((100,1)),x]
x_b
```

```python
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
```

```python
x_new = np.array([[0],[2]])
x_new
```

```python
x_new_b = np.c_[np.ones((2,1)), x_new]
y_predict = x_new_b.dot(theta_best)
```

```python
plt.plot(x,y,'b.')
plt.plot(x_new, y_predict,'r-')
plt.xlabel("$X_1$", fontsize = 20)
plt.ylabel('y', rotation =0, fontsize = 20)
plt.axis([0,2.5,0,15])
plt.show()
```

```python
def grad_descent_plot(eta, iteration, m):
    theta = np.random.randn(2,1)
    for iterations in range(iteration):
        gradients = 2/m * x_b.T.dot(x_b.dot(theta)- y )
        if iterations <10:
            y_predict = x_new_b.dot(theta)
            style = 'b-' if iteration > 0 else 'r--' 
            plt.plot(x_new, y_predict,style)
        
        theta = theta - eta* gradients
    plt.title(label = f'n={eta}')
    plt.plot(x,y,'b.')
    plt.xlabel("$X_1$", fontsize = 20)
    plt.ylabel('y', rotation =0, fontsize = 20)
    plt.axis([0,2.5,0,20])
```

```python
plt.figure(figsize=(12,4))
plt.subplot(131); grad_descent_plot(0.1, 1000, 100)
plt.subplot(132); grad_descent_plot(0.01, 1000, 100)
plt.subplot(133); grad_descent_plot(0.25, 1000,10)
plt.show()
```
