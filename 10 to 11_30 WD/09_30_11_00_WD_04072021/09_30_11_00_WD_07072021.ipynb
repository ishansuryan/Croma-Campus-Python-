```python
from sklearn.svm import LinearSVR
import numpy.random as rnd
import matplotlib.pyplot as plt
import numpy as np
rnd.seed(42)
m = 50
X = 2 * rnd.rand(m,1)
# print(X)
y = (4 + 3 * X + rnd.randn(m,1)).ravel()
# print(y)
plt.scatter(X,y)
plt.show()
```

```python
lin_svr = LinearSVR(epsilon=1.5)
lin_svr.fit(X,y)
```

```python
x_test = np.linspace(0,2,100).reshape(-1,1)
y_pred = lin_svr.predict(x_test)
y_top = y_pred + 1.5
y_bottom = y_pred -01.5
```

```python
plt.scatter(X,y)
plt.plot(x_test,y_pred,'g-')
plt.plot(x_test,y_top,'k--')
plt.plot(x_test,y_bottom,'k--')
plt.show()
```

```python
y_pred = lin_svr.predict(X)
supp_vec_X = X[np.abs(y-y_pred)>1.5]
supp_vec_y = y[np.abs(y-y_pred)>1.5]
plt.scatter(supp_vec_X,supp_vec_y)
plt.show()
```
