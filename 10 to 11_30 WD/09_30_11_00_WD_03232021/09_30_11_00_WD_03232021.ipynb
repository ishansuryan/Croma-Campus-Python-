```python
import numpy as np
from math import *
```

```python
x = np.linspace(-1000000,1000000,10000000)
```

```python
y = 1/(1+math.e**-x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,3))
plt.plot(x,y,'b-')
plt.grid(True)
plt.axis([-12,12,-1,1.5])
plt.show()
```

```python
x = np.random.rand(100)
x = list(x)
x
```

```python
x1,x2,y1,y2 = [],[],[],[]
for i in x:
    if i>=0.5:
        y1.append(-log(i))
        x1.append(i)
    else:
        y2.append(-log(1-i))
        x2.append(i)
```

```python
plt.figure(figsize = (6,6))
plt.plot(y1,x1,'b.')
plt.plot(y2,x2,'r.')
plt.show()
```
