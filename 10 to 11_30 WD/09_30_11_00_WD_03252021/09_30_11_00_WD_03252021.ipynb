```python
import numpy as np
```

```python
p_cap = np.linspace(0.001,1, 1000, endpoint=False)
```

```python
p_cap[999]
```

```python
c_1 = -np.log(p_cap)
c_0 = -np.log(1-p_cap)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(13,4))
plt.subplot(1,2,1)
plt.plot(p_cap,c_1,'g-')
plt.xlabel('probalility')
plt.ylabel('Cost Function')
plt.title('y = 1')
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(p_cap,c_0,'b-')
plt.xlabel('probalility')
plt.ylabel('Cost Function')
plt.title('y = 0')
plt.grid(True)
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
iris.keys()
```

```python
x = iris.data
y = iris.target
cat = iris.target_names
features = iris.feature_names
```

```python
import pandas as pd
pd.set_option('display.max_rows', 10)
```

```python
flower = pd.DataFrame(x, columns=features, index = [x_ for x_ in range(len(x))])
y =  pd.Series(cat[y])
```

```python
# flower.rename(columns = ({0:features[0],1:features[1],2:features[2],3:features[3]}), inplace = True)
```

```python
flower.head()
```

```python
np.random.seed(42)
shuffle_index = np.random.permutation(150)
shuffle_index
```

```python
flower = flower.iloc[shuffle_index]
y = y.iloc[shuffle_index]
flower
```

```python
y_verginica = (y == 'virginica')
y_verginica
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(flower,y_verginica,test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
log_clf = LogisticRegression(random_state=42)
```

```python
log_clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_train, log_clf.predict(x_train), output_dict = True)
report['True']
```

```python
x_petal_w = x_train.iloc[:,3:]
```

```python
log_clf.fit(x_petal_w,y_train)
```

```python
x_new = np.linspace(0,3,1000).reshape(-1,1)
pred = log_clf.predict_proba(x_new)
```

```python
plt.figure(figsize=(15,5))
plt.plot(x_new,pred[:,1:],'g-', label = 'Verginica')
plt.plot(x_new,pred[:,0],'b--', label = 'Not Verginica')
plt.legend()
plt.grid()
plt.xticks([x/10 for x in range(31)])
plt.show()
```
