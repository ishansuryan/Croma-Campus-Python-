```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_v3 import InceptionV3 , decode_predictions, preprocess_input
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = InceptionV3(weights = 'imagenet')
```

```python
path = '/home/athharv/Downloads/stairs.jpg'             # image path
path_1 = '/home/athharv/Downloads/mobile.jpg'
img = image.load_img(path, target_size = (299,299))     # importing image along resising image to required size
img = image.load_img(path_1, target_size = (299,299))
plt.imshow(np.asarray(img))                             # displaying the image
# plt.imshow(np.asarray(img_1))
plt.show()
```

```python
img_arr = image.img_to_array(img)               # conbvetring image to array
img_arr = np.array([img_arr])                   # making batch of images
img_processed = preprocess_input(img_arr)       # preprocessing the image
pred = model.predict(img_processed)             # taking prediction from model
```

```python
print(f'My prediction is {decode_predictions(pred, top= 3)}')
```
