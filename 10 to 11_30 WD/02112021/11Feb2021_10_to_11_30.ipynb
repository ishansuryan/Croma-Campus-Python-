```python
def line(x,c):
    m = .3
    y = m * x + c
    return y
```

```python
import matplotlib.pyplot as plt
```

```python
x_num = [1,32,36,85,5]
y_list = []

for i in x_num:
    z = line(i,0)
    y_list.append(z)
```

```python
plt.plot(x_num,y_list)
plt.yticks([0,5,10,15,20,25,30])
plt.show()
```
