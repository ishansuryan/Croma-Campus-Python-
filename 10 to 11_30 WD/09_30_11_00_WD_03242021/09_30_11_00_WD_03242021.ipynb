```python
import numpy as np
```

```python
p_cap = np.linspace(0.001,1, 1000, endpoint=False)
```

```python
p_cap[999]
```

```python
c_1 = -np.log(p_cap)
c_0 = -np.log(1-p_cap)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(13,4))
plt.subplot(1,2,1)
plt.plot(p_cap,c_1,'g-')
plt.xlabel('probalility')
plt.ylabel('Cost Function')
plt.title('y = 1')
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(p_cap,c_0,'b-')
plt.xlabel('probalility')
plt.ylabel('Cost Function')
plt.title('y = 0')
plt.grid(True)
```
