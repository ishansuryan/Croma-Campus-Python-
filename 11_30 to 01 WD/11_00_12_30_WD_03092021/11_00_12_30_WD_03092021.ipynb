```python
import numpy as np
import pandas as pd
```

```python
xs = np.array([0,1])
ys = np.array([0,1])
```

```python
X = {'X_cord': pd.Series(xs), 'y_cords': pd.Series(ys)}
```

```python
x_ = pd.DataFrame(X)
x_
```

```python
plt.plot(xs,ys,'ro')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss = 'hinge', penalty = 'l2')
y=[0,1]
clf.fit(x_,y)
```

```python
clf.coef_,clf.intercept_
```

```python
a = np.linspace(0,1,11)
```

```python
b = (-a*clf.coef_[0,0]-clf.intercept_[0])/clf.coef_[0,1]
```

```python
plt.plot(a,b,'r-', marker ='*')
plt.plot(xs,ys,'bo')
plt.grid(True)
plt.show()
```

```python
clf.predict([[2,2]])
```
