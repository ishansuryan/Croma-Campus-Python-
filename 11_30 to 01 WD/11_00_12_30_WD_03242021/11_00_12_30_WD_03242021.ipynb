```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
-np.log(1-.2)
```

```python
x = np.arange(0.0001,1,.0001)
len(x)
```

```python
c_1 = -np.log(x)
c_0 = -np.log(1-x)
```

```python
plt.figure(figsize = (13,5))
plt.subplot(1,2,1)
plt.plot(x,c_1,'b-')
plt.title('y=1')
plt.grid(True)
plt.xlabel('probability')
plt.ylabel('Cost function')
plt.subplot(1,2,2)
plt.plot(x,c_0,'g-')
plt.title('y=0')
plt.grid(True)
plt.xlabel('probability')
plt.ylabel('Cost function')
plt.xticks([x/10 for x in range(11)])
plt.yticks([x for x in range(10)])
plt.xlim(0,1)
plt.ylim(0,10)
plt.show()
```
