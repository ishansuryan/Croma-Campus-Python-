```python
import numpy as np
from math import *
```

```python
x = np.linspace(-1000000,1000000,10000000)
```

```python
y = 1/(1+e**-x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12,3))
plt.plot(x,y,'b-')
plt.axis([-12,12,-1,1.2])
plt.grid(True)
plt.show()
```

```python
x = np.arange(0.0001,1,.0001)
len(x)
```

# For class y = 1 cost function is = -log(p)

```python
cost_func = []
for i in x:
    cost_func.append(-log(i))
```

```python
plt.plot(x,cost_func,'r-')
plt.xlabel('-log(p)')
plt.ylabel('Cost function')
plt.title('For y=1')
plt.grid(True)
plt.axis([0,1,0,10])
plt.xticks([x/10 for x in range(11)])
plt.yticks([x for x in range(11)])
plt.show()
```

# For class y = 0 cost function is = -log(1-p)

```python
cost_func = []
for i in x:
    z = 1-i
    cost_func.append(-log(z))
```

```python
plt.plot(x,cost_func,'g-')
plt.xlabel('-log(p)')
plt.ylabel('Cost function')
plt.title('For y=0')
plt.grid(True)
plt.axis([0,1,0,10])
plt.xticks([x/10 for x in range(11)])
plt.yticks([x for x in range(11)])
plt.show()
```
