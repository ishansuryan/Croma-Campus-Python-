```python
from sklearn.svm import LinearSVR
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 2 * np.random.rand(50,1)
y = (4 + 3 * x + 2*np.random.rand(50,1)).ravel()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
lin_svr = LinearSVR(epsilon=01.8)
lin_svr.fit(x,y)
```

```python
x_test = np.linspace(0,2,100).reshape(-1,1)
y_pred = lin_svr.predict(x_test)
y_top = y_pred + 1.8
y_bottom = y_pred -1.8
plt.scatter(x,y)
plt.plot(x_test,y_pred,'g-')
plt.plot(x_test,y_top,'k--')
plt.plot(x_test,y_bottom,'k--')
plt.show()
```

```python
lin_svr.coef_, lin_svr.intercept_
```

```python
y_new = x_test*lin_svr.coef_+ lin_svr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x_test,y_new,'g-')
plt.show()
```
