```python
import numpy as np
np.random.seed(42)
```

```python
X = 2* np.random.rand(100,1)        # Uniformly distributed array with dimension (100,1)
```

```python
y = 4       +           3 * X       + np.random.randn(100,1)
#  theta 0          theta1                  gAUSSIAN nOISE
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,y,'b.')
plt.xlabel('$X_1$', fontsize = 20)
plt.ylabel('y', rotation = 0, fontsize = 20)
plt.axis([0,2.5,0,20])
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X]
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)        # putting the vsalues in formulae
```

```python
theta_best
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
X_new_b
```

```python
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new,y_predict, 'r-',linewidth = 2, label = "Predictions")
plt.plot(X,y,'b.')
plt.xlabel("$X_1$", fontsize = 20)
plt.ylabel('y', fontsize = 20)
plt.axis([0,2.5,0,15])
plt.legend(fontsize = 20, loc = 'upper left')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
eta = 0.5
n_iterations = 1000
m = 100

theta = np.random.randn(2,1)

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    if iteration <10:
        plt.plot(X_new_b, X_new_b.dot(theta),'r-')
    theta = theta - eta* gradients


plt.plot(X,y,'b.')
plt.xlabel("$X_1$", fontsize = 20)
plt.ylabel('y', fontsize = 20)
plt.axis([0,2.5,0,15])
# plt.legend(fontsize = 20, loc = 'upper left')
plt.show()
```
